```python
import smtplib
import ssl
from email.message import EmailMessage

def send_email(sender_email, receiver_email, subject, message, smtp_server, smtp_port, username, password):
    # Create the email container
    email = EmailMessage()
    email["From"] = sender_email
    email["To"] = receiver_email
    email["Subject"] = subject
    email.set_content(message)

    # Security context for the connection
    context = ssl.create_default_context()

    try:
        # Connect to the server and send
        with smtplib.SMTP_SSL(smtp_server, smtp_port, context=context) as server:
            server.login(username, password)
            server.send_message(email)
            print("Email sent successfully!")
    except Exception as e:
        print(f"Error: {e}")

# --- Configuration ---
sender_email = "shreya.pandey0520@gmail.com"
receiver_email = input("Enter the email id of the recipient: ")
subject = "Hello from Python!"
message = "This is a test email sent using Python."

# Gmail Settings
smtp_server = "smtp.gmail.com"
smtp_port = 465
username = "shreya.pandey0520@gmail.com"

# IMPORTANT: Use a Google 'App Password' here, NOT your regular login password
password = "xyxx wpdb ebbk dnnq"

# Execute
send_email(sender_email, receiver_email, subject, message, smtp_server, smtp_port, username, password)
```
